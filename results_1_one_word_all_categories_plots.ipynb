```python
import os
import PIL
from PIL import Image
from IPython.display import Image
from IPython.display import Image, display

# Load plots from folder.
base_path = os.getcwd() + '/plots/one_word_all_categories/'
file_names = [ base_path+f for f in os.listdir(base_path) if not '.ipynb' in f]

# Sort plots in order.
files_with_n = []
for f in file_names:
    n = f.split('/')[-1].split('_')[0]
    n = int(n)
    files_with_n.append((n,f))
    
files_sorted = sorted(files_with_n, key=lambda tup: tup[0])
files_sorted = sorted(files_with_n, key=lambda tup: tup[0])

print('Showing top {} words:\n\n\n'.format(len(files_sorted)))
show_N = 30
for i, f in files_sorted:
    if i > show_N:
        break
    name = f.split('/')[-1].split('_')[1].split('.')[0]
    print('{}. {}'.format(i, name))
    display(Image(filename=f))
```
